Alphavantage API
Pull data for major US market index and save to csv

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import time
import random


# Import API key
from api_keys import api_key
```

API code

```python
# function to run API request
def mi_api(cnt):
    MarketIndex=('DJIA', 'DXY', 'VIX')
    #('DJIA','DJT','DJU','NDX','NYA','SPX','MID','OEX','SOX','XAU','XOI','RUT')
    return MarketIndex[cnt]

# dictionary to hold the json result    

mainDict = {}
# setup URL
url = "https://www.alphavantage.co/query?"
function = "TIME_SERIES_MONTHLY"

# pull API:  note, this code will send an API request every minute.  
# the time to run this code depends on the number of Symbols (currently 12 symbols)
i = 0
oldtime = time.time()
print("API procedure star at: ", time.ctime())
while True:  
    if i < 3:
        Market = mi_api(i)
        query_url = f"{url}function={function}&symbol={Market}&apikey={api_key}"
        mainDict.update({Market : requests.get(query_url).json()}  )         
        #time.sleep(65)
        i += 1 
        print("API procedure complete at: ", time.ctime(), Market)
    if i == 3:  
        break
print("API procedures complete at: ", time.ctime())
```

Convert to dataframe

```python
#create a dataframe for dictionary
dataT=pd.DataFrame()
dataT['Symbol'] = ''
dataT['Week'] = ''
dataT['Open'] = ''
dataT['High'] = ''
dataT['Low'] = ''
dataT['Close'] = ''
dataT['Volume'] = ''

# unpack dictionary result from Json and add to dataframe
for key in mainDict:
    try:
        for lkey in mainDict[key]['Monthly Time Series']:
            dataT= dataT.append(pd.Series([key, lkey
              , mainDict[key]['Monthly Time Series'][lkey]['1. open']
              ,mainDict[key]['Monthly Time Series'][lkey]['2. high']
              ,mainDict[key]['Monthly Time Series'][lkey]['3. low']
              ,mainDict[key]['Monthly Time Series'][lkey]['4. close']
              ,mainDict[key]['Monthly Time Series'][lkey]['5. volume']
                                      ], index=dataT.columns), ignore_index=True)
        print('data added to dictionary for :', key)
            
    except (KeyError, IndexError):
        print('error data not added to dataframe: ', key)
        

#save into CSV
dataT.to_csv("output/MarketIndex.csv", index=False, header=True)

# view data in datframe
dataT.head()
```

```python
# turn to numeric, calculate % delta between open and close
dataT["Date"]=dataT['Week'].astype('datetime64[ns]')
dataT["Open"] = dataT["Open"].apply(pd.to_numeric, downcast='float', errors='coerce')
dataT["High"] = dataT["High"].apply(pd.to_numeric, downcast='float', errors='coerce')
dataT["Low"] = dataT["Low"].apply(pd.to_numeric, downcast='float', errors='coerce')
dataT["Close"] = dataT["Close"].apply(pd.to_numeric, downcast='float', errors='coerce')
dataT["Volume"] = dataT["Volume"].apply(pd.to_numeric, downcast='float', errors='coerce')
dataT["Delta"] = dataT["Close"]- dataT["Open"]
dataT["Percent Delta"] = dataT["Delta"]/ dataT["Open"]
   
dataT.head()
```

```python
# dataframe for % delta, per symbol, per month

pDelta = dataT.groupby(['Date', 'Symbol'])['Percent Delta'].sum().unstack(fill_value=0)
pDelta.head()
```

```python
#Plot
ax = pDelta.plot(lw=2, colormap='Paired', marker='.', markersize=1, figsize=(16,4),
         title='Percent Change in Price.')
ax.set_ylabel("Percent Change")
ax.set_xlabel("")
ax.grid(linestyle='-', linewidth=0.2)

# Save Figure
plt.savefig("output/MarketIndexTrend.png", pad_inches = 3, bbox_inches='tight')
```

CRYPTO CURRENCY DATA

```python
# function to run API request
def coin_api(cnt):
    MarketIndex=('LTC','BCH','ETC','DOGE','DASH','XMR','XRP', 'BTC')
    return MarketIndex[cnt]

    
# dictionary to hold the json result    

coinDict = {}
function = "DIGITAL_CURRENCY_MONTHLY"
Mar = 'USD'


i = 0
oldtime = time.time()
print("API procedure star at: ", time.ctime())
while True:  
    if i < 8:
        Market = coin_api(i)
        query_url = f"{url}function={function}&symbol={Market}&market={Mar}&apikey={api_key}"
        coinDict.update({Market : requests.get(query_url).json()}  )          
        if i ==4:
            time.sleep(65)
        i += 1 
        print("API procedure complete at: ", time.ctime(), Market)
    if i == 8:  
        break
print("API procedures complete at: ", time.ctime())
```

```python
#create a dataframe for dictionary
coinT=pd.DataFrame()
coinT['Symbol'] = ''
coinT['Week'] = ''
coinT['Open'] = ''
coinT['High'] = ''
coinT['Low'] = ''
coinT['Close'] = ''
coinT['Volume'] = ''

# unpack dictionary result from Json and add to dataframe
for key in coinDict:
    try:
        for lkey in coinDict[key]['Time Series (Digital Currency Monthly)']:
            coinT= coinT.append(pd.Series([key, lkey
              , coinDict[key]['Time Series (Digital Currency Monthly)'][lkey]['1a. open (USD)']
              ,coinDict[key]['Time Series (Digital Currency Monthly)'][lkey]['2a. high (USD)']
              ,coinDict[key]['Time Series (Digital Currency Monthly)'][lkey]['3a. low (USD)']
              ,coinDict[key]['Time Series (Digital Currency Monthly)'][lkey]['4a. close (USD)']
              ,coinDict[key]['Time Series (Digital Currency Monthly)'][lkey]['5. volume']
                                      ], index=coinT.columns), ignore_index=True)
        print('data added to dictionary for :', key)
            
    except (KeyError, IndexError):
        print('error data not added to dataframe: ', key)
        

#save into CSV
coinT.to_csv("output/CoinIndex.csv", index=False, header=True)       

# view data in datframe
coinT.head()
```

```python
# turn to numeric, calculate % delta between open and close
coinT["Date"]=coinT['Week'].astype('datetime64[ns]')
coinT["Open"] = coinT["Open"].apply(pd.to_numeric, downcast='float', errors='coerce')
coinT["High"] = coinT["High"].apply(pd.to_numeric, downcast='float', errors='coerce')
coinT["Low"] = coinT["Low"].apply(pd.to_numeric, downcast='float', errors='coerce')
coinT["Close"] = coinT["Close"].apply(pd.to_numeric, downcast='float', errors='coerce')
coinT["Volume"] = coinT["Volume"].apply(pd.to_numeric, downcast='float', errors='coerce')
coinT["Delta"] = coinT["Close"]- coinT["Open"]
coinT["Percent Delta"] = coinT["Delta"]/ coinT["Open"]
   
coinT.head()
```
