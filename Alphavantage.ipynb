Alphavantage API
Pull data for major US market index and save to csv

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
import time
import random


# Import API key
from api_keys import api_key
```

API code

```python
# function to run API request
def mi_api(cnt):
    MarketIndex=('DJIA','DJT','DJU','NDX','NYA','SPX','MID','OEX','SOX','XAU','XOI','RUT')
    return MarketIndex[cnt]

# dictionary to hold the json result    

mainDict = {}
# setup URL
url = "https://www.alphavantage.co/query?"
function = "TIME_SERIES_WEEKLY"

# pull API:  note, this code will send an API request every minute.  
# the time to run this code depends on the number of Symbols (currently 12 symbols)
i = 0
oldtime = time.time()
print("API procedure star at: ", time.ctime())
while True:  
    if i < 12:
        Market = mi_api(i)
        query_url = f"{url}function={function}&symbol={Market}&apikey={api_key}"
        mainDict.update({Market : requests.get(query_url).json()}  )         
        time.sleep(65)
        i += 1 
        print("API procedure complete at: ", time.ctime(), Market)
    if i == 12:  
        break
print("API procedures complete at: ", time.ctime())
```

Convert to dataframe

```python
#create a dataframe for dictionary
dataT=pd.DataFrame()
dataT['Symbol'] = ''
dataT['Week'] = ''
dataT['Open'] = ''
dataT['High'] = ''
dataT['Low'] = ''
dataT['Close'] = ''
dataT['Volume'] = ''

# unpack dictionary result from Json and add to dataframe
for key in mainDict:
    try:
        for lkey in mainDict[key]['Weekly Time Series']:
            dataT= dataT.append(pd.Series([key, lkey
              , mainDict[key]['Weekly Time Series'][lkey]['1. open']
              ,mainDict[key]['Weekly Time Series'][lkey]['2. high']
              ,mainDict[key]['Weekly Time Series'][lkey]['3. low']
              ,mainDict[key]['Weekly Time Series'][lkey]['4. close']
              ,mainDict[key]['Weekly Time Series'][lkey]['5. volume']
                                      ], index=dataT.columns), ignore_index=True)
        print('data added to dictionary for :', key)
            
    except (KeyError, IndexError):
        print('error data not added to dataframe: ', key)
        

#save into CSV
dataT.to_csv("output/MarketIndex.csv", index=False, header=True)

# view data in datframe
dataT.head()
```

```python
# turn to numeric, calculate % delta between open and close
dataT["Date"]=dataT['Week'].astype('datetime64[ns]')
dataT["Open"] = dataT["Open"].apply(pd.to_numeric, downcast='float', errors='coerce')
dataT["High"] = dataT["High"].apply(pd.to_numeric, downcast='float', errors='coerce')
dataT["Low"] = dataT["Low"].apply(pd.to_numeric, downcast='float', errors='coerce')
dataT["Close"] = dataT["Close"].apply(pd.to_numeric, downcast='float', errors='coerce')
dataT["Volume"] = dataT["Volume"].apply(pd.to_numeric, downcast='float', errors='coerce')
dataT["Delta"] = dataT["Close"]- dataT["Open"]
dataT["Percent Delta"] = dataT["Delta"]/ dataT["Open"]
   
dataT.head()
```

```python
# Obtain the x and y coordinates for each symbol:  'DJIA','DJT','DJU','NDX','NYA','SPX','MID','OEX','SOX','XAU','XOI','RUT'
DJIA = dataT[dataT['Symbol'] == 'DJIA']
DJT = dataT[dataT['Symbol'] == 'DJT']
DJU = dataT[dataT['Symbol'] == 'DJU']
NDX = dataT[dataT['Symbol'] == 'NDX']
NYA = dataT[dataT['Symbol'] == 'NYA']
SPX = dataT[dataT['Symbol'] == 'SPX']
MID = dataT[dataT['Symbol'] == 'MID']
OEX = dataT[dataT['Symbol'] == 'OEX']
SOX = dataT[dataT['Symbol'] == 'SOX']
XAU = dataT[dataT['Symbol'] == 'XAU']
XOI = dataT[dataT['Symbol'] == 'XOI']
RUT = dataT[dataT['Symbol'] == 'RUT']
```
